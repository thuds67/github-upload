```python
#DATA CLEANING
#Replacing missing values with the most common country
net['country'].value_counts()
```

```python
#After cleaning country its time to sort the the cells with multiple entries in country
net['new_country']=net['country'].apply(lambda x : x.split(',')[0])
net.drop('country', axis = 1, inplace = True)
```
